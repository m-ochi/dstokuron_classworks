## 散布図

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
symbols = []
for l in iris.target:
    if l == 0:
        s = "o"
    elif l == 1:
        s = ","
    else:
        s = "x"

    symbols.append(s)
```

```python
x = iris.data[:, 0]
y = iris.data[:, 1]
for i, data in enumerate(iris.data):
    x = data[0]
    y = data[1]
    l = iris.target[i]
    if l == 0:
        s = "o"
    elif l == 1:
        s = "+"
    else:
        s = "x"
        
    plt.plot(x, y, marker=s, color='b')
    
plt.title('Iris', size=16)
plt.xlabel(iris.feature_names[0], size=12)
plt.ylabel(iris.feature_names[1], size=12)
#plt.legend()
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame(iris.data, columns=["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"])
```

```python
#df
```

```python
tdf = pd.DataFrame(iris.target, columns=["target"])
tdf["target"] = tdf["target"].astype(str)
```

```python
df = pd.concat((df, tdf),axis=1)
```
